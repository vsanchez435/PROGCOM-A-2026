###Batalla por equipos
1. Calcular quienes son los que sacaron mas que el promedio, y quien fue la mayor nota y con cuanto. Considere que son 10 estudiantes

2. Imprimir los votos obtenidos por 5 candidatos distintos. Suponga que votan n usuarios.
Seleccione los dos primeros y vuelva a ejecutar la votación con solo esos dos (segunda vuelta).

````python
#1 BATALLA POR EQUIPOS
nombres = ["Ana", "Juan", "Pedro", "Maria", "Luis", "Elena", "Pablo", "Lucia", "Jose", "Rosa"]
notas = [85, 90, 70, 95, 80, 75, 100, 88, 60, 92]

promedio = sum(notas) / 10
print("El promedio de la clase es:", promedio)

print("Estudiantes por encima del promedio ")

for i in range(10):
    if notas[i] > promedio:
        print(nombres[i], "sacó", notas[i])

nota_mayor = max(notas)
posicion = notas.index(nota_mayor)
nombre_mayor = nombres[posicion]

print("El mejor fue", nombre_mayor, "con una nota de:", nota_mayor)
````

public class Main {
    public static void main(String[] args) {
        String[] nombres = {"Ana", "Juan", "Pedro", "Maria", "Luis", "Elena", "Pablo", "Lucia", "Jose", "Rosa"};
        int[] notas = {85, 90, 70, 95, 80, 75, 100, 88, 60, 92};


        double suma = 0;
        for (int nota : notas) {
            suma += nota;
        }
        double promedio = suma / notas.length;
        System.out.println("El promedio de la clase es: " + promedio);

        System.out.println("Estudiantes por encima del promedio:");
        for (int i = 0; i < nombres.length; i++) {
            if (notas[i] > promedio) {
                System.out.println(nombres[i] + " sacó " + notas[i]);
            }
        }

        int notaMayor = notas[0];
        int posicion = 0;

        for (int i = 1; i < notas.length; i++) {
            if (notas[i] > notaMayor) {
                notaMayor = notas[i];
                posicion = i;
            }
        }

        String nombreMayor = nombres[posicion];
        System.out.println("\nEl mejor fue " + nombreMayor + " con una nota de: " + notaMayor);
    }
}

````python
#2
n = int(input("Ingrese el número de votantes: "))

candidatos = ["abelardo", "cepeda", "paloma", "vicky", "botero"]
votos = [0, 0, 0, 0, 0]

print("Primera vuelta")
print("Opciones:")
for i in range(5):
    print(i+1, "-", candidatos[i])

for i in range(n):
    voto = int(input(f"Votante {i+1}, elija un candidato (1-5): "))
    if 1 <= voto <= 5:
        votos[voto-1] += 1
    else:
        print("Voto inválido")

print("Resultados de la primera vuelta:")
for i in range(5):
    print(candidatos[i], ":", votos[i], "votos")

max1 = 0
max2 = 1

if votos[max2] > votos[max1]:
    max1, max2 = max2, max1

for i in range(2, 5):
    if votos[i] > votos[max1]:
        max2 = max1
        max1 = i
    elif votos[i] > votos[max2]:
        max2 = i

print("Pasan a segunda vuelta:")
print(candidatos[max1])
print(candidatos[max2])

votos2 = [0, 0]

print("Segunda vuelta")
print("1 -", candidatos[max1])
print("2 -", candidatos[max2])

for i in range(n):
    voto = int(input(f"Votante {i+1}, elija (1-2): "))
    if voto == 1:
        votos2[0] += 1
    elif voto == 2:
        votos2[1] += 1
    else:
        print("Voto inválido")

print("Resultados de la segunda vuelta:")
print(candidatos[max1], ":", votos2[0], "votos")
print(candidatos[max2], ":", votos2[1], "votos")

if votos2[0] > votos2[1]:
    print("Ganador:", candidatos[max1])
elif votos2[1] > votos2[0]:
    print("Ganador:", candidatos[max2])
else:
    print("Empate en la segunda vuelta")
````

import java.util.Scanner;

public class Main {
    public static void main(String[] args) {

                System.out.println("Voto inválido");

````python
#3
print("MENU")
menu = {
    "1": ("Combo 1",        20000),
    "2": ("Combo 2",        25000),
    "3": ("Combo 3",        18000),
    "4": ("Combo 4",        24500),
    "5": ("Papas",           3000),
    "6": ("Jugos naturales", 7000),
    "7": ("Cerveza",         3500),
    "8": ("Agua",            4000),
    "9": ("Gaseosas",        2500),
}

total_dia = 0
ventas = {v[0]: 0 for v in menu.values()}

print("\nMENÚ:")
for k, (nombre, precio) in menu.items():
    print(f"  {k}. {nombre}: ${precio:,}")
print("  Escriba SALIR para cerrar caja")

continuar = 1

while continuar == 1:
    opcion = input("ORDEN DEL CLIENTE O ESCRIBA SALIR: ")

    if opcion == "SALIR":
        continuar = 0
    elif opcion in menu:
        nombre, precio = menu[opcion]
        total_dia += precio
        ventas[nombre] += 1
        print(f"{nombre} Total acumulado: ${total_dia:,}")
    else:
        print("Opción no válida")

print("RESUMEN DEL DÍA ")
for producto, cantidad in ventas.items():
    if cantidad > 0:
        precio = [v[1] for v in menu.values() if v[0] == producto][0]
        print(f"{producto}: {cantidad} unidades = ${cantidad * precio:,}")

print(f"TOTAL: ${total_dia:,}")
````

import java.util.HashMap;
import java.util.Map;
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);


        Map<String, Producto> menu = new HashMap<>();
        menu.put("1", new Producto("Combo 1", 20000));
        menu.put("2", new Producto("Combo 2", 25000));
        menu.put("3", new Producto("Combo 3", 18000));
        menu.put("4", new Producto("Combo 4", 24500));
        menu.put("5", new Producto("Papas", 3000));
        menu.put("6", new Producto("Jugos naturales", 7000));
        menu.put("7", new Producto("Cerveza", 3500));
        menu.put("8", new Producto("Agua", 4000));
        menu.put("9", new Producto("Gaseosas", 2500));


        Map<String, Integer> ventas = new HashMap<>();
        for (Producto p : menu.values()) {
            ventas.put(p.nombre, 0);
        }

        int totalDia = 0;


        System.out.println("--- MENÚ ---");
        for (String id : menu.keySet()) {
            System.out.println(id + ". " + menu.get(id).nombre + ": $" + menu.get(id).precio);
        }
        System.out.println("Escriba 'SALIR' para cerrar caja");


        boolean continuar = true;
        while (continuar) {
            System.out.print("\nORDEN DEL CLIENTE O ESCRIBA SALIR: ");
            String opcion = sc.nextLine().toUpperCase();

            if (opcion.equals("SALIR")) {
                continuar = false;
            } else if (menu.containsKey(opcion)) {
                Producto p = menu.get(opcion);
                totalDia += p.precio;
                
                ventas.put(p.nombre, ventas.get(p.nombre) + 1);
                
                System.out.println(p.nombre + " - Total acumulado: $" + totalDia);
            } else {
                System.out.println("Opción no válida");
            }
        }

      
        System.out.println("RESUMEN DEL DÍA ");
        for (String id : menu.keySet()) {
            Producto p = menu.get(id);
            int cantidad = ventas.get(p.nombre);
            if (cantidad > 0) {
                System.out.println(p.nombre + ": " + cantidad + " unidades = $" + (cantidad * p.precio));
            }
        }
        System.out.println("TOTAL FINAL: $" + totalDia);

        sc.close();
    }
}

class Producto {
    String nombre;
    int precio;

    Producto(String nombre, int precio) {
        this.nombre = nombre;
        this.precio = precio;
    }
}
