#20. Escribe un programa que adivine entre un conjunto de diez colores definidos por el usuario, con un máximo de 10 intentos.

```python
#forma 1
colores = ["rojo", "azul", "verde", "amarillo", "naranja", "morado", "rosa", "blanco", "negro", "gris"]
print(f"Piensa en uno de estos colores: {colores}\n")

for color in colores:
    respuesta = input(f"¿Tu color es el {color}? (si/no): ").strip().lower()

    if respuesta == "si":
        print("¡Genial! Adiviné tu color.")
        break
    else:
        print("Vaya, seguiré intentando...")

print("Fin del juego.")
```

```python
#forma 2
colores = ["rojo", "azul", "verde", "amarillo", "naranja", "morado", "rosa", "blanco", "negro", "gris"]
print(f"Piensa en uno de estos colores: {colores}\n")

for i in range(10):
    color_actual = colores[i]
    respuesta = input(f"Intento {i+1}: ¿Tu color es el {color_actual}? (si/no): ").strip().lower()

    if respuesta == "si":
        print("¡Genial! Adiviné tu color.")
        break
    else:
        print("Vaya, seguiré intentando...")

print("Fin del juego.")
```

```python
#codigo del ejercicio en java de la forma 3
import java.util.Scanner;

public class AdivinaColor {
    public static void main(String[] args) {
        Scanner teclado = new Scanner(System.in);

        String[] colores = {"rojo", "azul", "verde", "amarillo", "naranja", "morado", "rosa", "blanco", "negro", "gris"};

        System.out.println("Piensa en uno de estos colores: [rojo, azul, verde, amarillo, naranja, morado, rosa, blanco, negro, gris]\n");

        for (int i = 0; i < 10; i++) {
            String colorActual = colores[i];

            System.out.print("Intento " + (i + 1) + ": ¿Tu color es el " + colorActual + "? (si/no): ");

            String respuesta = teclado.nextLine().trim().toLowerCase();

            if (respuesta.equals("si")) {
                System.out.println("¡Genial! Adiviné tu color.");
                break;
            } else {
                System.out.println("Vaya, seguiré intentando...");
            }
        }

        System.out.println("Fin del juego.");

        teclado.close();
    }
}
```

#24. Desarrolla un programa que muestre el patrón de un triángulo con asteriscos.

```python
#forma 1
niveles = [1, 2, 3, 4, 5]
print("Triángulo rectangular\n ")
for cantidad in niveles:
    print("*" * cantidad)
```

```python
#forma 2
cantidad_filas = 5
print("Triángulo rectangular\n")

for i in range(cantidad_filas):
    print("*" * (i + 1))
```

```python
#codigo del ejercicio en java de la forma 1
import java.util.Scanner;

public class potencia {
    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.print("Ingrese la base: ");
        int base = sc.nextInt();

        System.out.print("Ingrese el exponente: ");
        int exponente = sc.nextInt();

        int resultado = 1;

        for (int i = 0; i < exponente; i++) {
            resultado = resultado * base;
        }

        System.out.println("La potencia es: " + resultado);
    }
}
```
