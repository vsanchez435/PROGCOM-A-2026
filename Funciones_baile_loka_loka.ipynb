```python
def paso1(a):
  for i in range(a):
    print("Codo flexionado de forma horizontal, moverlo a la recha la derecha junto con la derecha")
    print("Pausa 1 segundo")
def paso2(a,b):
  for i in range(b):
    print("subir el brazo derecho y mover la cadera la derecha")
    print("subir el brazo izquierdo y mover la cadera la izquierda")
def paso3(a,b,c):
  for i in range(c):
    print("saltar")
def paso4(a,b,c,d):
  for i in range(d):
    print("subir los dos brazos y mover de derecha a izquierda")
def paso5(a,b,c,d,e):
  for i in range(e):
    print("cerrar puños, flexionar codos y mover de derecha a izquierda ")
def paso6(a,b,c,d,e,f):
  for i in range(f):
    print("agarrarse de las manos y mover la cadera de derecha a izquierda ")
def paso7(a,b,c,d,e,f,g):
  for i in range(g):
    print("cerrar puños, flexionar codos y mover de derecha a izquierda ")
    print("cerrar puños, flexionar codos y levantar pierna derecha ")
    print("cerrar puños, flexionar codos y levantar pierna izquierda ")
def paso8(a,b,c,d,e,f,g,h):
  for i in range(h):
    print("tocar hombro derecho con la mano izquierda ")
    print("tocar hombro izquierdo con la mano derecha ")
    print("Unir las manos y ponerlas debajo de la cara ")
def baile_loka_loka(a,b,c,d,e,f,g,h,j):
  print("esperar unos segundos antes de iniciar el baile")
  for i in range(j):
    paso1(a)
    paso2(a,b)
    paso3(a,b,c)
    paso4(a,b,c,d)
    paso5(a,b,c,d,e)
    paso6(a,b,c,d,e,f)
    paso7(a,b,c,d,e,f,g)
    paso8(a,b,c,d,e,f,g,h)
def remate():
  print("Colocar mano izquierda en la cintura, levantar brazo derecho y flexionar piernas")
```

```python
baile_loka_loka(1,2,4,4,4,2,1,1,1)
remate()
```
