````python
numero = 3

print(f"TABLA DEL {numero} ")
for i in range(1, 11):
    resultado = numero * i
    suma = ""
    for j in range(i):
        suma = suma + str(numero)
        if j < i - 1:
           suma = suma + " + "
    print(f"{numero} x {i} = {suma} = {resultado}")
````
