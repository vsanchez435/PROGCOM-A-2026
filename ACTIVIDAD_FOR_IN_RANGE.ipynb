````python
#programa que calcule el factorial de un número. - FOR (VARIABLE) IN RANGE (CANTIDADD E VECES)
numero = int(input("Ingrese un numero: "))

factorial = 1

for i in range(1, numero + 1):
    if numero < 0:
        print("No existen factoriales para números negativos")
        break
    factorial = factorial * i

if numero == 0:
    print("El factorial de 0 es 1")
elif numero > 0:
    print("El factorial de", numero, "es:", factorial)
````

````python
#programa que calcule el factorial de un número. - FOR VARIABLE IN RANGE (VI, VF+1,P)
numero = int(input("Ingrese un numero: "))

factorial = 1

for i in range(1, numero + 1, 1):   # VI=1, VF=numero, P=1
    if numero < 0:
        print("No existen factoriales para números negativos")
        break
    factorial = factorial * i

if numero == 0:
    print("El factorial de 0 es 1")
elif numero > 0:
    print("El factorial de", numero, "es:", factorial)
````

```
import java.util.Scanner;
public class Main
{
	public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);

        System.out.print("Escribe un número para sumar sus dígitos: ");
        String numeroUsuario = scanner.nextLine();
        
        int suma = 0;

        for (int i = 0; i < numeroUsuario.length(); i++) {
          
            char caracter = numeroUsuario.charAt(i);
          
            suma += Character.getNumericValue(caracter);
        }

        System.out.println("Resultado: " + suma);
        
        scanner.close();
	}
}
```

````python
#programa que determine si un número es primo.- FOR (VARIABLE) IN (ALGO)
numero = int(input("Ingrese un número: "))

for i in range(numero):
    if numero <= 1:
        print("No es un número primo")
        break
    elif i >= 2 and numero % i == 0:
        print("No es un número primo")
        break
else:
    print("Es un número primo")
````

````python
#programa que determine si un número es primo. - FOR (VARIABLE) IN RANGE (CANTIDAD DE VECES)
numero = int(input("Ingrese un número: "))

for i in range(2, numero):
    if numero <= 1:
        print("No es un número primo")
        break
    elif numero % i == 0:
        print("No es un número primo")
        break
else:
    if numero > 1:
        print("Es un número primo")
````

````python
#programa que determine si un número es primo. - FOR VARIABLE IN RANGE (VI, VF+1,P)
numero = int(input("Ingrese un número: "))

for i in range(1, numero + 1, 1):  # VI=1, VF=numero, P=1
    if numero <= 1:
        print("No es un número primo")
        break
    elif i > 1 and i < numero and numero % i == 0:
        print("No es un número primo")
        break
else:
    print("Es un número primo")
````

```
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.print("Ingrese la base: ");
        int base = sc.nextInt();

        System.out.print("Ingrese el exponente: ");
        int exponente = sc.nextInt();

        int resultado = 1;

        for (int i = 0; i < exponente; i++) {
            resultado = resultado * base;
        }

        System.out.println("La potencia es: " + resultado);
    }
}
   
```



````python
#programa que calcule la potencia de un número sin usar el operador **.- FOR (VARIABLE) IN RANGE (CANTIDAD DE VECES)
base = int(input("Ingrese la base: "))
exponente = int(input("Ingrese el exponente: "))
resultado = 1

for i in range(exponente):
    resultado = resultado * base
print("La potencia es:", resultado)
````
