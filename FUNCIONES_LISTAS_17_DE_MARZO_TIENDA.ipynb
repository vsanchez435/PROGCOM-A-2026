 Es un pedazo de codigo que se ejecuta aparte. A veces se conoce como copiado dinámico.

 Ventajas de implmentar las funciones:

 1. Vuelve el código mas limpio
 2. Elimina el código espaguetti
 3. Establece un orden de búsqueda
 5. Permite reutilizacion del código



```
def nombre_de_la_funcion(parametros):
  Aqui va todo lo que dese ejecutare
  return algo
```

Existen 4 tipos de funciones:

1. Sin parametros, sin retorno
2. Sin parametro, con retorno
3. Con parametros, sin retorno
4. Con parametros, con retornos
