##**Ejercicio**5
Crea un diccionario con nombres de estudiantes y sus notas.
Muestra el promedio de todos los estudiantes y de un estudiante específico, empleando dos funciones.

image.png

`````python
estudiantes = {
    "Ana": [5, 4, 3],
    "Luis": [3, 2, 1],
    "Marta": [5, 5, 5]
    }
`````

`````python
def promedio_individual(nombre):
    notas = estudiantes[nombre]
    resultado = sum(notas) / len(notas)
    return resultado
`````

`````python
def promedio_total():
    suma_notas = sum(estudiantes.values())
    cantidad = len(estudiantes)
    promedio = suma_notas / cantidad
    print(f"El promedio de la clase es: {promedio}")
`````
