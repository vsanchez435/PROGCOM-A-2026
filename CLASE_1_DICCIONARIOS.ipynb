`````python
#Listas
galletas=["angelitos","diablitos","dinosaurio"]
galletas_precio=[20000,15000,15000]
galletas_cant=[3,5,10]
`````

`````python
def menu_galletas():
  for i in range(3):
    print(f"{galletas[i]}:{galletas_precio[i]}, cantidad:{galletas_cant[i]}")
`````

`````python
menu_galletas()
`````
