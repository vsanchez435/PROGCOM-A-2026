````python
import random

jugando = True

while jugando:
    eleccion = input("Elige entre 1. Piedra, 2. Papel, 3. Tijera: ")

    while eleccion != "1" and eleccion !="2" and eleccion !="3":
      eleccion=input("Valor no valido")

    yo = int(eleccion)
    m = random.randint(1, 3)
    print(f"La máquina eligió: {m}")

    if yo == m:
        print("empate. Ambos elegimos 🪨")
    elif yo == 1 and m == 2:
        print("Perdiste, La maquina eligio 2")
    elif yo == 2 and m == 3:
        print("Perdiste, La maquina eligio")
    elif yo == 3 and m == 1:
        print("Perdiste, La maquina eligio")
    else:
        print("Ganaste")

    continuar = input("¿Quieres seguir jugando? (si/no): ")
    if continuar == "no":
        jugando = False
        print("Gracias por jugar. ¡Hasta la próxima!")
````
