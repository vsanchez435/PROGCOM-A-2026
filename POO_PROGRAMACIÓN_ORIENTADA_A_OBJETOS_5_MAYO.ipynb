````python
class Edificio():
  #Construyendo el constructor
  def __init__(self,material, ubicación, fachada, ventanas, puertas):

    self.material=material
    self.ubicación=ubicación
    self.fachada=fachada
    self.ventanas=ventanas
    self.puertas=puertas

  def direccion(self,tipo):
    return f"Tipo de edificación:{tipo}\nDirección:{self.ubicación}"

  def descripcion(self, tipo):
    print(f" {self.nombre} es un{tipo} ubicado en {self.ubicación}")
    print(f" Tiene {self.ventanas} ventana(s), {self.puertas} puertas")
    print(f" Su fachada es {self.fachada} ")
````
