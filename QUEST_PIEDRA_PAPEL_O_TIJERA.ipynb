```python
import random
nombre = input("Hola, ¿cómo te llamas?: ")
intentos = int(input(f"Mucho gusto {nombre}. ¿Cuántas rondas quieres jugar?: "))

mis_puntos = 0
puntos_maquina = 0

for i in range(intentos):
    print(f"Ronda {i+1}")

    eleccion = int(input("Escribe el número de tu opción 1. Piedra ✊,2. Papel ✋ o 3. Tijera ✌️ : "))

    m = random.randint(1, 3)

    print(f"{nombre} eligió: {eleccion}")
    print(f"La máquina eligió: {m}")

    if eleccion == m:
        print("¡Es un empate! 🤝")

    elif (eleccion == 1 and m == 3) or (eleccion == 2 and m == 1) or (eleccion == 3 and m == 2):
        print(f"¡Punto para {nombre}! 🎉")
        mis_puntos = mis_puntos + 1

    else:
        print("Punto para la máquina 😢")
        puntos_maquina = puntos_maquina + 1

print(f"Marcador final: {nombre} {mis_puntos} - {puntos_maquina} Máquina")

if mis_puntos > puntos_maquina:
    print(f"¡Felicidades {nombre}, ganaste el torneo! 🏆")
elif mis_puntos < puntos_maquina:
    print("La próxima vez será, la máquina ganó. 🤖")
else:
    print("¡Increíble! Terminaron en empate total. 😮")
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import random

puntos_usuario = 0
puntos_maquina = 0

# 2. Pedir el nombre antes de mostrar la interfaz
nombre = input("¿Cómo te llamas? ")

instrucciones = widgets.HTML(value=f"<h3>¡Hola {nombre}! Elige tu opción:</h3>")
resultado_pantalla = widgets.HTML(value="<b>Esperando tu jugada...</b>")
marcador_pantalla = widgets.HTML(value=f"<b>Marcador: {nombre} 0 - 0 Máquina</b>")

def jugar(boton_presionado):
    global puntos_usuario, puntos_maquina

    eleccion = boton_presionado.description

    opciones = ["Piedra ✊", "Papel ✋", "Tijera ✌️"]
    maquina = random.choice(opciones)

    if eleccion == maquina:
        mensaje = f"Empate. Ambos eligieron {eleccion} 🤝"
    elif (eleccion == "Piedra ✊" and maquina == "Tijera ✌️") or \
         (eleccion == "Papel ✋" and maquina == "Piedra ✊") or \
         (eleccion == "Tijera ✌️" and maquina == "Papel ✋"):
        mensaje = f"¡Ganaste! {eleccion} vence a {maquina} 🎉"
        puntos_usuario = puntos_usuario + 1
    else:
        mensaje = f"Perdiste. {maquina} vence a {eleccion} 😢"
        puntos_maquina = puntos_maquina + 1

    resultado_pantalla.value = f"<h4>{mensaje}</h4>"
    marcador_pantalla.value = f"<b>Marcador: {nombre} {puntos_usuario} - {puntos_maquina} Máquina</b>"

btn_piedra = widgets.Button(description="Piedra ✊", button_style='info')
btn_papel = widgets.Button(description="Papel ✋", button_style='info')
btn_tijera = widgets.Button(description="Tijera ✌️", button_style='info')

btn_piedra.on_click(jugar)
btn_papel.on_click(jugar)
btn_tijera.on_click(jugar)

caja_botones = widgets.HBox([btn_piedra, btn_papel, btn_tijera])
display(instrucciones, caja_botones, resultado_pantalla, marcador_pantalla)
```
