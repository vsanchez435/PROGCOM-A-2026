## 1. IF /  ELSE
Solo tiene dos posibles respuestas (V) o (F).

 Sintaxis

```
if (condición):
  aqui se ejecuta codigo ***si*** la condicion es verdadera

else:
   aqui se ejecuta el codigo ***si** la condicion es falsa
```

````python
temp=float(input("Temperatura registrada:"))
if temp>=27: print("Póngase algo fresco👚☀️")
elif temp>=20 and temp<27: print("Abrígate🧥🌥️")
elif temp>=16 and temp<20: print("Abrígate más 🧣🧥🌦️")
else: print("Está helando🥶🥶")
````

##SI ternario (Easy hand if)
Se emplea cuando la estructura solo posee un if-else (recomendado).

sintaxis


```
variable = (acción si la condición es verdadera) if (condición) else (acción si la condición es falsa)
```

Sintaxis en java


```
variable = (condicion)? (acción si la condición es verdadera): (acción si la condición es falsa)
```
