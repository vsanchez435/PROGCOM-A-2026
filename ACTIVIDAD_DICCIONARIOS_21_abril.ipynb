##**Ejercicio**5
Crea un diccionario con nombres de estudiantes y sus notas.
Muestra el promedio de todos los estudiantes y de un estudiante específico, empleando dos funciones.

image.png

````python
estudiantes = {
    "Ana": [5, 4, 3],
    "Luis": [3, 2, 1],
    "Marta": [5, 5, 5]
    }
````

````python
def promedio_individual(nombre):
    notas = estudiantes[nombre]
    resultado = sum(notas) / len(notas)
    return resultado
````

````python
print(f"El promedio de Ana es: {promedio_individual('Ana')}")
````

````python
def promedio_total():

    suma_de_promedios = promedio_individual("Ana") + promedio_individual("Luis") + promedio_individual("Marta")

    promedio_total = suma_de_promedios / 3
    print(f"El promedio de la clase es: {promedio_total}")
    return promedio_total
````
