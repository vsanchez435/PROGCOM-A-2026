###GQ3 - PRODUCTIVIDAD LABORAL

```python
#CALCULAR PRODUCTIVIDAD LABORAL
print("SISTEMA DE CONTROL DE PRODUCCIÓN")

producto = input("¿Qué producto se fabricó hoy?: ")
unidades = float(input(f"¿Cuántas unidades de {producto} se produjeron?: "))
operarios = int(input("¿Cuántos operarios trabajaron?: "))
horasoperario = float(input("¿Cuántas horas trabajó cada operario?: "))

totalhoras = operarios * horasoperario

productividad = unidades / totalhoras

print(f"RESUMEN DE PRODUCTIVIDAD PARA: {producto}")
print(f"Total de Horas: {totalhoras} hrs")
print(f"Productividad: {round(productividad, 2)} unidades por hora-hombre")

if productividad > 5:
    print("¡Excelente! El equipo fue muy eficiente.")
else:
    print("Atención: La productividad está por debajo de la meta.")
```
